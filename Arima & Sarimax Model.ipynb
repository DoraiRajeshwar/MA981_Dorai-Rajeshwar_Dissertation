```python
def adf_test(series):
    result=adfuller(series)
    print('ADF Statistics: {}'.format(result[0]))
    print('p- value: {}'.format(result[1]))
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis, reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
```

```python
adf_test(df_close['Close'])
```

```python
# Since the time series is non-stationary, it implies that the statistical properties, such as the mean and variance, are not constant over time. Non-stationary time series can exhibit trends, seasonality, or other patterns, making them more challenging to analyze using traditional statistical methods.

# To make the time series stationary, you might consider applying transformation techniques like differencing, logarithmic transformation, or seasonal decomposition to remove trends and seasonality. The goal of making the series stationary is to achieve a constant mean and variance over time, which is essential for many time series analyses and forecasting models.
```

```python
# Fit the ARIMA model
```

```python
from statsmodels.tsa.arima.model import ARIMA
import statsmodels.api as sm
import numpy as np


arima_predict = []
arima_model = ARIMA(train['Close'],order=(5,2,0)).fit()
forecast_values= arima_model.forecast(96)
arima_predict.extend(forecast_values)
arima_predict = np.array(arima_predict).reshape((96,))
test['arima_pred'] = arima_predict
arima_model.summary()
```

```python
print((test))
```

```python
#from sklearn.metrics import mean_squared_error
#from sklearn.metrics import r2_score
#import math
#errors = mean_squared_error(test['Close'], test['arima_pred'])
#print('Mean Squared Error: %.3f' % errors)
#print('Root Mean Squared Error: %.3f' % math.sqrt(errors))
#print(r2_score(test['Close'], test['arima_pred']))

from sklearn.metrics import mean_absolute_error
# Evaluating the model using MAE Evaluation Metric
print(mean_absolute_error(test['Close'], test['arima_pred']))
from sklearn.metrics import mean_squared_error
print(mean_squared_error(test['Close'], test['arima_pred']))
##R score also known as the coefficient of determination gives the measure of how good a model fits to a given dataset. 
##It indicates how closer are the predicted values to the actual values.

from sklearn.metrics import r2_score
print(r2_score(test['Close'], test['arima_pred']))
```
